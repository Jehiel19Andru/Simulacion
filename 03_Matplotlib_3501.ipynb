## Representación gráfico de Datos.

Si a la función de trazado se le da una matriz de datos, la usará como coordenadas en el eje vertical, y utilizará el índeice de cada punto de datos en el array como la coordenada horizontal.

```python
plt.plot([1, 2, 5, 7, 8, 3, 1])
plt.show()
```

Se muestra la grafica de acuerdo a los puntos dados.

Tambien se puede proporcionar dos matrices: una para el eje horizontal, y otra para el eje vertical.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

### Pueden modificarse las longitudes de los ejes para que la figura no se vea tan ajustada.

```python
plt.plot([-3, -1, 0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10]) # [xmin, xmax, ymin, ymax]
plt.show()
```

Se sigue el mismo procedimiento para pintar una función matematica

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2

plt.plot(x, y)
plt.show()
```

Tambien pueden modificarse el estilo de la gráfica para que contenga más información.

```python
plt.plot(x, y)
plt.title("Square Fucntion")
plt.xlabel("Chido")
plt.ylabel("y =**2")
plt.grid(True)
plt.show()
```

Pueden ponerse gráficas y cambiar el estilo de las funciones.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y, 'b--', x, y2, 'g') #Esto quiere decir que marcar una linea azul punteada
plt.show()
```

```python
# Separando en diferentes lineas las funciones.

import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square Fucntion")
plt.xlabel("Chido")
plt.ylabel("y =**2")
plt.grid(True)

plt.plot(x, y, 'b--')
plt.plot(x, y2, 'g') 
plt.show()
```

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.title("Square Fucntion")
plt.xlabel("Chido")
plt.ylabel("y =**2")
plt.grid(True)

plt.plot(x, y, 'b--', label = "x**2")
plt.plot(x, y2, 'g', label = "x + 1") 
plt.legend( loc = "best") # La situa en la mejor localización.
plt.show()
```

Tambien pueden crearse dos gráficas que no se superpongan, estas gráficas se organizan 
en un grid y se denominan subplots.

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1

plt.subplot(1, 2, 1) # 1 Rows, 2 Columns, 1st subplot
plt.plot(x, y, 'b--')

plt.subplot(1, 2, 2) # 1 Rows, 2 Colums, 2nd Subplot
plt.plot(x, y2, 'g')

plt.show()
```

Para que las gráficas no queden tas ajustadas, se puede hacer la figura más grande.

```python
plt.figure(figsize = (14, 6))

plt.subplot(1, 2, 1) # 1 Rows, 2 Columns, 1st subplot
plt.plot(x, y, 'b--')
plt.title("Square Fucntion", fontsize = 14)
plt.xlabel("Chido", fontsize = 14)
plt.ylabel("y =**2", fontsize = 14)
plt.grid(True)

plt.subplot(1, 2, 2) # 1 Rows, 2 Colums, 2nd Subplot
plt.plot(x, y2, 'g')
plt.title("Lineal Function", fontsize = 14)
plt.xlabel("Chido 2", fontsize = 14)
plt.ylabel("y =**2", fontsize = 14)
plt.grid(True)

plt.show()
```

### Scatter Plots

```python
from numpy.random import rand
x, y = rand(2, 100)
plt.scatter(x, y)
plt.show()
```

```python
from numpy.random import rand
x, y = rand(2, 100)
x2, y2 = rand(2, 100)
plt.scatter(x, y, c = 'red')
plt.scatter(x2, y2, c = 'blue')

plt.show()
```

## Histogramas

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3.3, 3]
plt.subplot(211)
plt.hist(data, bins = 10, rwidth = 0.8)

plt.subplot(212)
plt.hist(data, bins = [1, 1.5, 2, 2.5, 3], rwidth = 0.95) 
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.show()
```
